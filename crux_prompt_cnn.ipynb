```python
import pandas as pd
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import BatchNormalization
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
from keras.utils.np_utils import normalize, to_categorical
```

```python
#for testing all files
x = np.loadtxt('extracted_final_output.csv', delimiter = ',')
y = np.loadtxt('extracted_final_label.csv', delimiter = ',')
```

```python
x = x.reshape(len(x), 190, 4, 1)
y = y.reshape(len(y), 1)
```

```python
train_raw, test_raw, target, target_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=0)
```

```python
train = normalize(train_raw, axis=1)
test = normalize(test_raw, axis=1)
target = to_categorical(target)
target_test = to_categorical(target_test)
```

```python
train.shape
```

```python
#plotting
# idx2 = random.randint(0, len(train))
for plot in range(len(target)):
    if target[plot][2] == 1:
        to_plot = train[plot]
        x_coordinate = [i for i in range(len(to_plot))]
        graphs = [];
        for time in to_plot:
            for count,channel in enumerate(time):
                if count >= len(graphs):
                    graphs.append([])
                graphs[count].append(channel[0])
        for graph in graphs:
            plt.plot(x_coordinate, graph)
        plt.ylim([-1,1])
    #         plt.title(target[plot])
        plt.show()
    #     print(target[idx2])
```

```python
model = Sequential()

model.add(Conv2D(32, (3,3), activation = 'sigmoid', input_shape = (190,4,1)))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())
model.add(Dense(64, activation = 'sigmoid'))
model.add(Dense(3, activation = 'sigmoid'))

model.compile(optimizer = 'rmsprop',loss = 'categorical_crossentropy', metrics = ['accuracy'])
#model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
history = model.fit(train, target, epochs = 2000, batch_size = 64)

#plot the training and validation accuracy and loss at each epoch
loss = history.history['loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'y', label='Training loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#plotting
# idx2 = random.randint(0, len(train))
for plot in range(len(test)):
    if target_test[plot][2] == 1:
        to_plot = test[plot]
        x_coordinate = [i for i in range(len(to_plot))]
        graphs = [];
        for time in to_plot:
            for count,channel in enumerate(time):
                if count >= len(graphs):
                    graphs.append([])
                graphs[count].append(channel[0])
        for graph in graphs:
            plt.plot(x_coordinate, graph)
        y_pred = model.predict(test[plot].reshape(1, 190, 4, 1))
        plt.ylim([-1,1])
        if np.argmax(y_pred) == 0:   
            plt.title("Blank with certainty " + str(y_pred[0][0]))
        if np.argmax(y_pred) == 1:   
            plt.title("Raise with certainty " + str(y_pred[0][1]))
        if np.argmax(y_pred) == 2:   
            plt.title("Clench with certainty " + str(y_pred[0][2]))
        plt.show()
    #     print(target[idx2])
```
